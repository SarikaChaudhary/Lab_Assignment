# Importing packages

```python
import credentials
import json
import csv
import pandas as pd
import tweepy
from tweepy.auth import OAuthHandler
from tweepy import API
from tweepy import OAuthHandler
```

# Tweepy class 
#Initialization of Tweepy keys
#Authentication 
#Init_name to initialize name the Twitter Account ID and no of tweets to fetch
#fetching tweets using fetch_tweets
#Saving the tweets after processing and extracting relevant fields to a csv file

```python
class Tweepy:
    def __init__(self):
        # Initializing key values using an external credentials.py file with Api_Credentials dictionary  
        self.__consumer_key =credentials.Api_Credentials['consumer_key']
        self.__consumer_secret = credentials.Api_Credentials['consumer_secret']
        self.__access_token = credentials.Api_Credentials['access_token']
        self.__access_token_secret = credentials.Api_Credentials['access_token_secret']
        
        self.__IDname="@midasIIITD"       # fetch tweets from @midasIIITD account
        self.__no_of_tweets=200           # fetch 200 tweets
    
    def authentication(self):
        # Creating the authentication object
        auth = tweepy.OAuthHandler(self.__consumer_key,self.__consumer_secret)
        # Setting your access token and secret
        auth.set_access_token(self.__access_token,self.__access_token_secret)
        # Creating the API object while passing in auth information
        api = tweepy.API(auth)
        return api
    
    
    def Init_name(self,name,tc):
        self.__TwitterIDname=name                       # to change the Twitter AccountID
        self.__no_of_tweets=tc                          # to change the no of tweets to be fetched
        
    def fetch_tweets(self,api):
        # fetching first 200 tweets from @midasIIITD twitter account with full_text extended version
        result= api.user_timeline(self.__IDname,count=self.__no_of_tweets,tweet_mode="extended")
        
        with open("JSONlines.json",'a',encoding='utf-8') as f_out:
            for status in result:
                json.dump(status._json, f_out)         # dumping json of each tweet in JSONlines.json file
                f_out.write('\n')
            f_out.close()
        print ('JSONlines file has been created!')
        return True
    
    
    def Save_toCSV(self):
        tweets_data=[]
        tweets_file = open("JSONlines.json", "r")
        for line in tweets_file:
            try:
                tweet = json.loads(line)
                tweets_data.append(tweet)
            except:
                continue

        tweets = pd.DataFrame()
        #extracting text of the tweet
        tweets['text'] = list([tweet['full_text'] for tweet in tweets_data])
        #extracting date and time of the tweet
        tweets['created_at'] = list([tweet['created_at']  for tweet in tweets_data])
        # number of retweets
        tweets['retweet_count'] = list([tweet['retweet_count']for tweet in tweets_data])
        # number of favourite/likes
        tweets['favorite_count'] = list([tweet['favorite_count'] for tweet in tweets_data])
        # number of images uploaded with tweet if any or None 
        tweets['No_of_Images'] = list([len(tweet['extended_entities']['media']) if 'extended_entities' in tweet else 'None' for tweet in tweets_data])

        # creating a csv file
        tweets.to_csv("Sarika_Chaudhary.csv",sep=',',encoding='utf-8')
        tweets_file.close()
        
        print ('csv has been created!')
        return True
```

#  Main function 

```python
def main():
    obj=Tweepy()               # object of Tweepy class
    api=obj.authentication()   # authentication
    obj.fetch_tweets(api)      # fetching tweets in the form of json and creating a JSONlines.json file
    obj.Save_toCSV()           # creating a csv file 
    
if __name__ == '__main__':
    main()
```
